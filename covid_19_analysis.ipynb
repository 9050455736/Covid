##  find the distribution of each state in % wrt total cases

```python
## NOTE:: SINCE THE CATEGORIES ARE MORE, SO PIE CHART BECOMES VERY MESSY
```

```python
top_5=df.groupby(["State/UTs"])["Total Cases"].agg(["mean"]).sort_values(by=["mean"],ascending=False).head(5)
```

```python
top_5
```

```python
bottom_5=df.groupby(["State/UTs"])["Total Cases"].agg(["mean"]).sort_values(by=["mean"],ascending=False).tail(5)
```

```python
bottom_5
```

```python
s=top_5["mean"]*100/df["Total Cases"].sum()
```

```python
s
```

```python
s1=bottom_5["mean"]*100/df["Total Cases"].sum()
```

```python
len(s1)
```

```python
len(s)
```

```python
plt.figure(figsize=(18,12))
plt.subplot(1,2,1)
plt.title("top 5 affected states",fontsize=20)
plt.bar((top_5["mean"]*100/df["Total Cases"].sum()).index,(top_5["mean"]*100/df["Total Cases"].sum()).values,color="red")
plt.xticks(rotation=90,fontsize=20)
plt.xlabel("State/UTs",fontsize=20)
plt.ylabel("% of total cases",fontsize=20)
for i in range(len(s)):
    plt.text(i,s.iloc[i],f"{round(s.iloc[i],3)}%",ha="center",fontsize=18)
plt.subplot(1,2,2)
plt.title("bottom 5 affected states",fontsize=20)
plt.bar((bottom_5["mean"]*100/df["Total Cases"].sum()).index,(bottom_5["mean"]*100/df["Total Cases"].sum()).values,color="green")
plt.xticks(rotation=90,fontsize=20)
plt.xlabel("State/UTs",fontsize=20)
plt.ylabel("% of total cases",fontsize=20)
for i in range(len(s1)):
    plt.text(i,s1.iloc[i],f"{round(s1.iloc[i],3)}%",ha="center",fontsize=18)
```

## segregating states on the basis of lockdown status

- Lockdown 1: wear mask
- lockdown 2: wear mask and only essential services
- lockdown 3: complete lockdown

```python
df.head(2)
```

```python
df[df["State/UTs"]=="Delhi"]# Total cases are 1434780
```

```python
df[df["State/UTs"]=="Jammu and Kashmir"]# Total cases are 317761
```

```python
df["Lockdown_Status"]=df["Total Cases"].apply(lambda x: " Lockdown 3" if x>=1434780 else ("Lockdown 2" if ((x<1434780) & (x>=317761)) else "Lockdown 1"))
```

```python
df.sample(10)
```

## now we find % of states having lockdowns

```python
df["Lockdown_Status"].value_counts()
```

```python
plt.pie(df["Lockdown_Status"].value_counts(),labels=df["Lockdown_Status"].value_counts().index,autopct="%.2f%%");
```

# states with highest active and death cases

```python
df.head(2)
```

```python
top_ac=df.groupby(["State/UTs"])["Active"].agg(["mean"]).sort_values(by=["mean"],ascending=False).head(5)
```

```python
top_ac
```

```python
ac=top_ac["mean"]*100/df["Active"].sum()
```

```python
top_d=df.groupby(["State/UTs"])["Deaths"].agg(["mean"]).sort_values(by=["mean"],ascending=False).head(5)
```

```python
top_d
```

```python
d=top_d["mean"]*100/df["Deaths"].sum()
```

```python
plt.figure(figsize=(18,12))
plt.subplot(1,2,1)
plt.title("states with highest active cases",fontsize=20)
plt.bar((top_ac["mean"]*100/df["Total Cases"].sum()).index,(top_ac["mean"]*100/df["Active"].sum()).values,color="blue")
plt.xticks(rotation=90,fontsize=20)
plt.xlabel("State/UTs",fontsize=20)
plt.ylabel("% of total active cases",fontsize=20)
for i in range(len(ac)):
    plt.text(i,ac.iloc[i],f"{round(ac.iloc[i],3)}%",ha="center",fontsize=18)
plt.subplot(1,2,2)
plt.title("states with highest death cases",fontsize=20)
plt.bar((top_d["mean"]*100/df["Total Cases"].sum()).index,(top_d["mean"]*100/df["Deaths"].sum()).values,color="red")
plt.xticks(rotation=90,fontsize=20)
plt.xlabel("State/UTs",fontsize=20)
plt.ylabel("% of total Death cases",fontsize=20)
for i in range(len(d)):
    plt.text(i,d.iloc[i],f"{round(d.iloc[i],3)}%",ha="center",fontsize=18)
```

## Insight:: 
- Maharashtra recorded highest active cases and highest deaths as well
- Delhi and Uttar Pradesh have recorded very high death cases despite not having very high active cases.
